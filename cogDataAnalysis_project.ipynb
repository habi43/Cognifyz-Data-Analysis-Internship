```python
# import necessary labriries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# load the Dataset
df = pd.read_csv("Dataset.csv")
#show the first 5 rows
df.head()
```

```python
# count the occurrences of each cuisine
cuisine_counts = df['Cuisines'].value_counts()
#get the top 3 cuisines
top_3_cuisines = cuisine_counts[:3]
#covert to Dataframe to display
top_3_cuisines_df = pd.DataFrame({'Cuisine': top_3_cuisines.index, 'Count': top_3_cuisines.values})
# display 5 rows
display(top_3_cuisines_df)
```

```python
#Count restaurants in each price range
price_counts = df ['Price range'].value_counts()
# conv to Dataframe
price_df = pd.DataFrame({'Price Range': price_counts.index, 'Count': price_counts.values})
# diplay the results
display(price_df)
```

```python
# to find the restaurants with most votes and least votes
top_voted = df.loc[df['Votes'].idxmax(), ['Restaurant Name', 'Votes']]
least_voted = df.loc[df['Votes'].idxmin(), ['Restaurant Name', 'Votes']]
# convert to Dataframe
votes_df = pd.DataFrame([top_voted, least_voted])
#display the table
display(votes_df)
```

```python
# count online delivery & table booking by price range
price_service_counts = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].value_counts()
# conv to Datafrme
price_service_df = price_service_counts.reset_index()
# display the table
display(price_service_df)
```

```python
# to save my analysis to excel 
top_3_cuisines_df.to_excel("Cuisine_Analysis.xlsx", index=False)
price_df.to_excel("Price_Analysis.xlsx", index=False)
votes_df.to_excel("Votes_Analysis.xlsx", index=False)
price_service_df.to_excel("Price_Service_Analysis.xlsx", index=False)
```
